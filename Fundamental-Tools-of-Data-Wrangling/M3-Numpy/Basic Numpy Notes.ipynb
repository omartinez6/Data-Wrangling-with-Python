## Creation

```python
ls = [1,2,3,4,5]
np.array(ls)
```

```python
ar1 = np.arange(10)
print('ar1:',ar1)

ar2 = np.arange(5,11)
print('ar2:',ar2)

ar3 = np.arange(0,13,3)
print('ar3:',ar3)

ar4 = np.arange(4,-4,-1)
print('ar4:',ar4)
```

```python
ar5 = np.arange(1,5,1)
print('ar5:',ar5,type(ar5[0]))

ar6 = np.arange(1,5,0.5)
print('ar6',ar6,type(ar6[0]))

ar7 = np.arange(1,5,dtype=float)
print('ar7:',ar7,type(ar7[0]))
```

## Getting basic information about and array

```python
ar = np.array([1,2,3,4,5])
array_type = type(ar)
array_size = np.size(ar)
array_dimensions = ar.ndim
array_data_type = ar.dtype

print(f'''Object type: {array_type},
Array size: {array_size},
Array dimensions: {array_dimensions},
Array data type: {array_data_type}''')
```

## Similar to lists

```python
ls = [x ** 2 for x in range(10)]
ar = np.arange(10) ** 2
print(ls, ar)
```

```python
print(ls[0], ar[0])
```

```python
print(ls[:3], ar[:3])
```

## Differences from lists

```python
ls = [1,2,3,4,5]
print(ls)
print(ls + ls)
print(ls * 2)
```

```python
ar = np.arange(1,6,1)
print(ar)
print(ar + ar)
print(ar * 2)
```

```python
ar1 = np.arange(1,6,1)
s = 2
ls = [2,2,2,2,2]
ar2 = np.array([2,2,2,2,2])

print(ar1 * s)
print(ar1 * ls)
print(ar1 * ar2)
```

```python
ar = np.arange(1,6,1)
print(ar + 2)
print(ar - 2)
print(ar / 2)
print(ar ** 2)
print(ar ** 0.5)
print(ar % 2)
```

## Multiplying two ndarrays

```python
ar1 = np.arange(1,10)
ar2 = np.arange(10,1,-1)
print(ar1, ar2, np.multiply(ar1,ar2),end='\n')
```

```python
print(ar1.dot(ar2))
```

```python
print(ar1@ar2)
```

## Some examples of universal functions

```python
ar_ints = np.arange(1,6)
ar_floats = np.arange(1,7.5,1.5)
```

```python
print(np.add(ar_ints, ar_floats))
```

```python
print(np.negative(ar_ints))
```

```python
print(np.square(ar_floats))
```

```python
print(np.hypot(ar_ints, ar_floats))
```

```python
print(np.greater(ar_ints, ar_floats))
```

```python
print(np.ceil(ar_floats))
```
