## Using boolean arrays

```python
import numpy as np
```

```python
np.random.seed(1)
ar = np.random.randint(100,size=10)
print(ar)
```

```python
odd_mask = ar % 2 == 1
print(odd_mask)
```

```python
print(ar[odd_mask])
```

```python
ar = np.random.randint(100, size=(10,10))
print(ar)
```

```python
print(ar[ar % 2 == 0])
```

```python
print(ar[ar < 50])
```

```python
ar2 = np.random.randint(100, size = (10,10))
print(ar2)
```

```python
print(ar[ar < ar2])
```

```python
print(ar[ar + ar2 > 100])
```
