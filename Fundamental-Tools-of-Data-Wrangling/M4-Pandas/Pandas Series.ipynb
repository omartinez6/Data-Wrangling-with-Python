# Pandas

```python
import numpy as np
import pandas as pd
```

## Creation

### Create series from ndarray

```python
s1 = pd.Series(np.arange(0,5))
print(s1)
```

### Create series with index

```python
s2 = pd.Series(np.arange(0,5), index = ['a','b','c','d','e'])
print(s2)
```

### Assign index to existing series

```python
s2.index = ['A','B','C','D','E']
print(s2)
```

```python
print(s2.index)
```

### Create one-item series from scalar

```python
s3 = pd.Series(5)
print(s3)
```

### Create series from list

```python
s4 = pd.Series([1,2,3,4,5])
print(s4)
```

```python
print(s4.index)
```

### Create series from dictionary

### Manipulation

```python
grd1 = pd.Series([17,44,28,8,3], index = ['A','B','C','D','F'])
grd2 = pd.Series([76,122,151,21,0], index = ['D','C','B','A','F'])
```

```python
print(grd1)
```

```python
print(grd2)
```

```python
grd_all = grd1 + grd2
print(grd_all)
```

```python
grd1 = pd.Series([17,44,28,8,3], index = ['A','B','C','D','F'])
grd2 = pd.Series([76,122,151,21], index = ['D','C','B','A'])
grd_all = grd1 + grd2
print(grd1, grd2, grd_all)
```

```python
grd_all = grd1.add(grd2, fill_value = 10)
print(grd_all)
```

### Comparing series

```python
print(type(dies1 == dies2))
```

```python
print(dies1[dies1 == dies2])
```

### Element-wise operations

```python
np.random.seed(1)
exam_grades = pd.Series(np.random.randint(60, 101,100))
print(exam_grades)
```

```python
curved_grades = exam_grades.multiply(1.05)
print(curved_grades)
```

```python
def convert_to_letter(grade):
    if grade >= 90:
        return 'A'
    elif grade >= 80:
        return 'B'
    elif grade >= 70:
        return 'C'
    elif grade >= 65:
        return 'D'
    else:
        return 'F'
```

```python
letter_grades = curved_grades.apply(convert_to_letter)
print(letter_grades)
```

```python
part_letter_grades = curved_grades.iloc[:50].apply(convert_to_letter)
print(part_letter_grades)
```
