## Arithmatic Operations

### Task: Ask the user to type in two positive int, and compute the result of (+, -, *, /, //, **, %)

```python
x = int(input('Ingresa un número positivo entero: '))
y = int(input('Ingresa otro número positivo entero: '))

suma = x + y
print(suma)

resta = x - y
print(resta)

multiplicacion = x * y
print(multiplicacion)

division = x / y
print(division)

division_entera = x // y
print(division_entera)

exponenciacion = x ** y
print(exponenciacion)

modulo = x % y
print(modulo)
```

## Relational Operations

### Task: Ask the user to type in two numbers, and compute the result of (<, <=, ==, >=, >, !=)

```python
x = int(input('Ingresa un número positivo entero: '))
y = int(input('Ingresa otro número positivo entero: '))

menor = x < y
print(menor)

menor_igual = x <= y
print(menor_igual)

igualdad = x == y
print(igualdad)

mayor_igual = x >= y
print(mayor_igual)

mayor = x > y
print(mayor)

diferente = x != y
print(diferente)
```

## Branching

### Task: What day is today?

Write a program that asks the user for a number in the range of 1 through 7. The program should display the corresponding day of the week, where

* 1    = Monday,
* 2    = Tuesday,
* 3    = Wednesday,
* 4    = Thursday,
* 5    = Friday,
* 6    = Saturday, and
* 7    = Sunday
* all other value entered = ERROR

```python
dia = int(input('Ingresa un número entero entre el 1 y el 7: '))

if dia == 1:
    print('Monday')
elif dia == 2:
    print('Tuesday')
elif dia == 3:
    print('Wednesday')
elif dia == 4:
    print('Thursday')
elif dia == 5:
    print('Friday')
elif dia == 6:
    print('Saturday')
elif dia == 7:
    print('Sunday')
else:
    print('ERROR')
```

## Repetition

### Task: Keep saving until you can retire.

Write a program to simulate how to save money for retirement.

1.  Asks the user for a number representing a new positive deposit for the retirement fund.
1.  Print out the amount of the deposit as well as the balance for each transaction.
1.  When the balance is equal or more than 1 million, print a message "You can retire now" to the user.
1.  Print "Congratualtions!" and exit the program.

```python
ahorros_totales = 0

while True:
    ahorro = float(input('Ingresa la cantidad a ahorrar: '))
    ahorros_totales = ahorros_totales + ahorro
    print(f'Haz ahorrado {ahorros_totales} dólares en total')
    if ahorros_totales > 1000000:
        print('You can retire now')
        print('Congratulations!')
        break
```
